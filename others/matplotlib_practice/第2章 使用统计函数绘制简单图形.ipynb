###  使用统计函数绘制简单图形
#### 函数bar()——用于绘制柱状图

```python
# -*- coding:utf-8 -*-

import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

# some simple data
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 1, 4, 5, 8, 9, 7, 2]

# create bar
# align='center'代表x刻度中的q、a、c...居于柱状中间，而不是边缘
plt.bar(x, y,align='center', color = "c", tick_label = ["q", "a", "c", "e", "r", "j", "b", "p"], hatch = "/")

#set x, y_axis label
plt.xlabel("箱子编号")
plt.ylabel("箱子重量（kg）")
plt.show()
```

#### 函数barh()——用于绘制条形图

```python
# -*- coding:utf-8 -*-

import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

# some simple data
x = [1, 2, 3, 4, 5, 6, 7, 8]
y = [3, 1, 4, 5, 8, 9, 7, 2]

# create bar
# align='center'代表x刻度中的q、a、c...居于柱状中间，而不是边缘
plt.barh(x, y,align='center', color = "c", tick_label = ["q", "a", "c", "e", "r", "j", "b", "p"], hatch = "/")

#set x, y_axis label
plt.xlabel("箱子重量（kg）")
plt.ylabel("箱子编号")
plt.show()
```

#### 函数hist()——用于绘制直方图

```python
# -*- coding:utf-8 -*-

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

# set test scores
boxWeight = np.random.randint(0, 10, 100)

x = boxWeight

# plot histogram
# bins 指条形的数量
bins = range(0, 11, 1)

plt.hist(x, bins = bins,
        color = "g",
        histtype = "bar",
        rwidth = 1, 
        alpha = 0.6)


#set x, y_axis label
plt.xlabel("箱子重量（kg）")
plt.ylabel("箱子数量（个）")

plt.show()
```

#### 函数pie()——用于绘制饼图

```python
# -*- coding:utf-8 -*-

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams["font.sans-serif"] = ["SimHei"]
mpl.rcParams["axes.unicode_minus"] = False

kinds = ["简易箱", "保温箱", "行李箱", "密封箱"]

colors = ["r", "g", "c", "b"]

soldNums = [0.05, 0.45, 0.15, 0.35]

#pie chart
plt.pie(soldNums,
       labels = kinds,
       autopct = "%3.1f%%", #应该是百分比
       startangle = 60, #开始的角度
       colors = colors)

plt.title("不同类型箱子的销售数量占比")

plt.show()
```

#### 函数polar()——用于绘制极线图

```python
# -*- coding:utf-8 -*-

import matplotlib.pyplot as plt
import numpy as np

barSlices = 12

theta = np.linspace(0.0, 2*np.pi, barSlices, endpoint=False)
r = 30*np.random.rand(barSlices)

plt.polar(theta,               # theta：每个标记所在射线与极径的夹角
          r,                   # 每个标记到原点的距离
         color = "chartreuse", # 指连线色是chartreuse
         linewidth = 2,
         marker = "*",
         mfc = "b",            # 指marker color是blue
         ms = 10               # 指marker size 是10
         )

plt.show()
```

#### 函数scatter()——用于绘制气泡图

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

a = np.random.randn(100)
b = np.random.randn(100)

# colormap:RdYlBu
plt.scatter(a, b, 
           s = np.power(10*a + 20*b, 2), # s: 散点标记的大小
           c = np.random.rand(100),      # c: 散点标记的颜色
           cmap = mpl.cm.RdYlBu,         # cmap:将浮点数映射成颜色映射表
           marker = "o")

plt.show()
```

#### 函数stem()——用于绘制棉棒图

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.5, 2*np.pi, 20)
y = np.random.randn(20)

plt.stem(x,                # 指定棉棒的x轴基线上的位置
         y,                # 绘制棉棒的长度
         linefmt = "-.",   # 棉棒的样式
         markerfmt = "o",  # 棉棒末端的样式
         basefmt = "-")    # 指定基线的样式

plt.show()
```

#### 函数boxplot()——用于绘制箱线图

```python
# -*- coding:utf-8 -*-

import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

mpl.rcParams["font.sans-serif"] = ["FangSong"]
mpl.rcParams["axes.unicode_minus"] = False

x = np.random.randn(1000)

plt.boxplot(x)  # x:绘制箱线图的输入数据

plt.xticks([1], ["随机数生成器AlphaRM"])  #将刻度1换成文字
plt.ylabel("随机数值")
plt.title("随机数生成器抗干扰能力的稳定性")

plt.grid(axis="y", ls = ":", lw = 1, color = "gray", alpha = 0.4)
plt.show()
```

#### 函数errorbar()——用于绘制误差棒图

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0.1, 0.6, 6)
y = np.exp(x)

plt.errorbar(x,          # 数据点的水平位置
             y,          # 数据点的*直位置
             fmt="bo:",  
             yerr=0.2,   # yerr: y轴方向的数据点的误差计算方法
             xerr=0.02)  # xerr: x轴方向的数据点的误差计算方法

plt.xlim(0, 0.7)

plt.show()
```
